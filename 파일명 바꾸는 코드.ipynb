## train 폴더

```python
'''ad'''

import os
import shutil

dir_name = 'picture' # 디렉터리 이름 변경

u = "C:\\Users\\USER\\Desktop\\binary_cnn_majority\\train\\ad"
q = u + '/' + dir_name + '_n'
p = u + '/' + dir_name

os.mkdir(q)
file_list = os.scandir(p)

num = 1

for file in file_list :
    n = file.name
    print(n, num)
    
    file_oldname = p + '\\' + n
 
    os.rename(file_oldname, q +'\\' + str(num) + '.jpg')
    
    num += 1

if(os.path.getsize(p) == 0):
    os.rmdir(p)

os.rename(q, p)
```
